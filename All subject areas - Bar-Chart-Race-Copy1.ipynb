```python
# import OS module
import os
 
# Get the list of all files and directories
path = "D:\COGNOS METRIX\DATA\All subject areas\DATA"
subject_areas_list = os.listdir(path)
 
print("Files and directories in '", path, "' :")
 
# prints all files
subject_areas_list
```

```python
for subject in subject_areas_list:
# create an Empty DataFrame object
    df = pd.DataFrame()

    for year in range(1996,2022):
        #Import data
        data = pd.read_excel(f"../DATA/All subject areas/DATA/{subject}/scimagojr country rank {year}.xlsx")[['Country','Documents']]
        data = data.rename(columns={"Documents": f"{year}"})
        data = data.set_index('Country')

        # Make a copy of data
        df = pd.concat([df, data], axis=1)
        
    df = df.fillna(0).astype(int).reset_index()
    
    df.to_excel(f'../DATA/All subject areas/subject_areas/{subject}.xlsx')

df
```

```python
import os, glob
import pandas as pd

path = "D:\COGNOS METRIX\DATA\All subject areas\subject_areas"

all_files = glob.glob(os.path.join(path, "*.xlsx"))

all_df = []
for f in all_files:
    df = pd.read_excel(f)#.reset_index()
    df['subject areas'] = f.split('\\')[-1].split('.')[0]
    all_df.append(df)
    
All_subject_areas = pd.concat(all_df, sort=True).reset_index().drop(['Unnamed: 0','index'], axis=1);

All_subject_areas['subject areas'] = All_subject_areas['subject areas'].str.replace(r'_', ' ')

All_subject_areas.to_excel(f'../DATA/All subject areas/All subject areas.xlsx')

All_subject_areas
```

```python
subject_areas_list = pd.DataFrame(All_subject_areas['subject areas'].unique())
subject_areas_list.to_excel(f'../DATA/All subject areas/subject_areas_list.xlsx')
subject_areas_list
```

```python
subject_areas_and_field = pd.read_excel(f"../DATA/All subject areas/subject_areas_and_field.xlsx").dropna()#.set_index('subject categories')
subject_areas_and_field 
```

```python
All_subject_areas = pd.merge(All_subject_areas, subject_areas_and_field, how="inner", on = 'subject areas')
All_subject_areas
```

```python
country_list = list(All_subject_areas['Country'].unique())
country_list
```

```python
for country in country_list:
    country_subject_areas = All_subject_areas.loc[All_subject_areas['Country'] == country]
    
    country_subject_areas.to_excel(f'../DATA/All subject areas/0country_subject_areas/{country}_subject_areas.xlsx')

country_subject_areas
```
